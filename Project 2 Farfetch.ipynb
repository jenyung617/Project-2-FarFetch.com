```python
from selenium import webdriver
import time
import re
import pandas as pd
import csv 

from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(ChromeDriverManager().install())

csv_file = open('Farfetch.csv', 'w', encoding='utf-8', newline='')
writer = csv.writer(csv_file)

urls = ['https://www.farfetch.com/shopping/women/clothing-1/items.aspx?page={}&view=180'.format(i) for i in range(1,3)]
index = 0

# a list to store all the prices
price_list = []
designer_list = []
description_list = []
season_list = []
for url in urls:
    index += 1
    print('Selenium on page {}'.format(index))
    driver.get(url)
    time.sleep(2)

    try:
        designers = driver.find_elements_by_xpath('//*[@data-test="productDesignerName"]')
        print(len(designers))
        descriptions = driver.find_elements_by_xpath('//*[@data-test="productDescription"]')
        print(len(descriptions))
        prices = driver.find_elements_by_xpath('//*[@data-test="price"]')
        print(len(prices))
        seasons = driver.find_elements_by_xpath('//*[@data-test="productMerchandiseLabel"]')
        print(len(seasons))
        for designer in designers:
            designer_list.append(designer.text) 
        for description in descriptions:
            description_list.append(description.text)             
        for price in prices[1:len(prices)]:
            price_list.append(price.text) 
        for season in seasons:
            season_list.append(season.text)             
    except:
        print('Elements cannot be found.')

#print(price_list) 
#print(designer_list)
#print(description_list)
#print(season_list)
        
# price is a list of webelement for one page only
# to access the actual price string, you need to iterate  thru the price list using for loop
# store all the price string to a list

clothing = {}
clothing['Designer'] = designer_list
clothing['Description'] = description_list
clothing['Price'] = price_list
clothing['Season'] = season_list
# writer.writerow(clothing.values())

# print(dict)
# csv_file.close()
# driver.close()
```
